# **Diabetes data**

````python
df3 = pd.read_csv("/content/drive/MyDrive/Assessment/diabetes_model.csv")
df3.head()
````

````python
plt.ylabel("Glucose Level")
plt.title("Glucose Quartile Deviation")
plt.boxplot(df3["Glucose"])
````

````python
plt.xlabel("Blood Pressure")
plt.ylabel("No. of peoples")
plt.title("Blood Pressure of peoples")
plt.hist(df3["BloodPressure"],linewidth=0.5, edgecolor="white")
````

````python
plt.bar(df3['Age'],df3['Pregnancies'],color='skyblue', width=2, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Pregnancies')
plt.title('Pregnancies per Age')
plt.show()
````

````python
plt.scatter(df3['Age'],df3['Pregnancies'])
plt.xlabel('Age')
plt.ylabel('Pregnancies')
plt.title('Pregnancies per age')
plt.xticks(rotation=90)
plt.show()
````

````python
fig,ax=plt.subplots(1,2,figsize=(20,8))

ax[0].scatter(df3['Age'],df3['Insulin'])
ax[0].set_title('Insulin - Scatter Plot')

ax[1].bar(df3['Age'],df3['BMI'])
ax[1].set_title('BMI - Bar Plot')

plt.show()
````

````python
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
````

````python
X = df3['Age']
Y = df3['BMI']
Z = df3['Glucose']

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(X, Y, Z, c=df3['BloodPressure'], cmap='viridis')

ax.set_xlabel('Age')
ax.set_ylabel('BMI')
ax.set_zlabel('Glucose')
ax.set_title('3D Scatter Plot of Age, BMI, and Glucose with Blood Pressure')

cbar = plt.colorbar(scatter)
cbar.set_label('BloodPressure')

plt.show()
````
